```python
import pandas as pd

# Load and combine JMP files
water = pd.read_csv("Basic and safely managed drinking water services.csv")
sanitation = pd.read_csv("Basic and safely managed sanitation services.csv")
defecation = pd.read_csv("Open defecation.csv")

jmp = pd.concat([water, sanitation, defecation], ignore_index=True)

# Clean
jmp.dropna(subset=['Country', 'Year', 'Numeric'], inplace=True)
jmp['Year'] = jmp['Year'].astype(int)

# Save full clean JMP
jmp.to_csv("jmp_clean.csv", index=False)
print("JMP clean:", jmp.shape)
print("Indicators:", jmp['Indicator'].unique())

# Clean Unified dataset - keep only relevant columns
unified = pd.read_csv("UnifiedDataset.csv")
cols_needed = ['Country', 'Year', 'GDP per Capita', 'Total Population',
               'Basic Drinking Water Services', 'Basic Sanization Services Total',
               'Safely Sanitation Total', 'Safely Sanitation Urban', 
               'Safely Sanitation Rural', 'Unsafe Wash Mortality Rate']

unified_clean = unified[cols_needed].dropna(subset=['Country', 'Year', 'GDP per Capita'])
unified_clean.to_csv("unified_clean.csv", index=False)
print("Unified clean:", unified_clean.shape)
```

```python
# Chart 1 - Line Chart data
# Use safely managed drinking water, national level only
chart1 = jmp[
    (jmp['Indicator'] == 'Population using safely managed drinking-water services (%)') &
    (jmp['Residence Area Type'] == 'National')
][['Country', 'WHO region', 'Year', 'Numeric']].copy()

chart1.rename(columns={'Numeric': 'water_access', 'WHO region': 'region'}, inplace=True)
chart1.to_csv("chart1_data.csv", index=False)
print("Chart 1 rows:", len(chart1))
print("Countries:", chart1['Country'].nunique())

# Chart 2 - Bar Chart data (same data, will filter by year in D3)
chart2 = chart1.copy()
chart2.to_csv("chart2_data.csv", index=False)
print("\nChart 2 rows:", len(chart2))
print("Sample:\n", chart1.head())
```
